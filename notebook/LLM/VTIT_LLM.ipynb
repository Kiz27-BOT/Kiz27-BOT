````python
# ==========================================
# PHẦN 1: KẾT NỐI DRIVE & CÀI ĐẶT THƯ VIỆN
# ==========================================
# 1. Mount Google Drive vào hệ thống Colab
from google.colab import drive
drive.mount('/content/drive')

# 2. Cài đặt các thư viện đọc đa định dạng (PyMuPDF cho PDF, python-docx cho Word)
!pip install pymupdf python-docx pandas -q

# ==========================================
# PHẦN 2: CODE ĐỌC SÂU TẤT CẢ FILE TRONG THƯ MỤC
# ==========================================
import os
import fitz  # PyMuPDF
import pandas as pd
import docx

# THAY ĐỔI ĐƯỜNG DẪN NẾU THƯ MỤC NẰM Ở VỊ TRÍ KHÁC
# Mặc định thư mục nằm ngoài cùng của MyDrive
THU_MUC_GOC = '/content/drive/MyDrive/AIRace_Track_Chatbot_LLM'

def doc_file_pdf(file_path):
    """Đọc text từ file PDF"""
    text = ""
    try:
        with fitz.open(file_path) as doc:
            for page in doc:
                text += page.get_text() + "\n"
    except Exception as e:
        print(f"Lỗi đọc PDF {file_path}: {e}")
    return text

def doc_file_docx(file_path):
    """Đọc text từ file Word"""
    text = ""
    try:
        doc = docx.Document(file_path)
        for para in doc.paragraphs:
            text += para.text + "\n"
    except Exception as e:
        print(f"Lỗi đọc DOCX {file_path}: {e}")
    return text

def doc_file_csv(file_path):
    """Đọc text từ file CSV / Excel"""
    text = ""
    try:
        df = pd.read_csv(file_path)
        text = df.to_string()
    except Exception as e:
        print(f"Lỗi đọc CSV {file_path}: {e}")
    return text

def doc_file_txt(file_path):
    """Đọc text từ file TXT, MD, JSON"""
    text = ""
    try:
        with open(file_path, 'r', encoding='utf-8') as f:
            text = f.read()
    except Exception as e:
        print(f"Lỗi đọc TXT {file_path}: {e}")
    return text

def quet_va_doc_thu_muc(thu_muc):
    """Duyệt đệ quy toàn bộ thư mục và đọc file"""
    if not os.path.exists(thu_muc):
        print(f"Không tìm thấy thư mục: {thu_muc}")
        print("Vui lòng kiểm tra lại đường dẫn xem thư mục có nằm trong thư mục con nào khác không.")
        return []

    kho_du_lieu = []

    print(f"Đang quét thư mục: {thu_muc}\n" + "="*50)

    # os.walk giúp quét sâu vào TẤT CẢ các thư mục con bên trong
    for root, dirs, files in os.walk(thu_muc):
        for file in files:
            file_path = os.path.join(root, file)
            ext = os.path.splitext(file)[1].lower()
            noidung = ""

            print(f"Đang đọc: {file}...")

            # Phân loại định dạng để xử lý
            if ext == '.pdf':
                noidung = doc_file_pdf(file_path)
            elif ext in ['.docx', '.doc']:
                noidung = doc_file_docx(file_path)
            elif ext == '.csv':
                noidung = doc_file_csv(file_path)
            elif ext in ['.txt', '.md', '.json']:
                noidung = doc_file_txt(file_path)
            else:
                print(f"  -> Bỏ qua định dạng: {ext}")
                continue

            if noidung.strip():
                kho_du_lieu.append({
                    "ten_file": file,
                    "thu_muc_chua": root,
                    "duong_dan": file_path,
                    "noi_dung": noidung.strip()
                })

    print("="*50)
    print(f"Đã quét và trích xuất thành công {len(kho_du_lieu)} file hợp lệ.")
    return kho_du_lieu

# ==========================================
# PHẦN 3: THỰC THI VÀ LƯU TRỮ
# ==========================================
# Chạy hàm quét
data_chatbot = quet_va_doc_thu_muc(THU_MUC_GOC)

# Kiểm tra dữ liệu thu được
if data_chatbot:
    print(f"\nVí dụ dữ liệu lấy được từ file đầu tiên ({data_chatbot[0]['ten_file']}):")
    print("-" * 50)
    # In ra 500 ký tự đầu tiên
    print(data_chatbot[0]['noi_dung'][:500] + "...\n")

    # (Tùy chọn) Lưu toàn bộ dữ liệu ra một file tổng hợp dạng JSON/CSV để dễ dàng nạp vào VectorDB sau này
    import json
    with open('/content/tong_hop_data_chatbot.json', 'w', encoding='utf-8') as f:
        json.dump(data_chatbot, f, ensure_ascii=False, indent=4)
    print("Đã lưu toàn bộ nội dung trích xuất vào file: /content/tong_hop_data_chatbot.json")
````

````python
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document

print("Bat dau xu ly PDF...")
pdf_files = glob.glob(f"{PDF_DIR}/*.pdf")
documents = []

# Đọc checkpoint để lấy danh sách các file đã xử lý xong
processed_files = []
if os.path.exists(CHECKPOINT_FILE):
    try:
        with open(CHECKPOINT_FILE, 'r') as f:
            processed_files = json.load(f)
    except:
        processed_files = []

print(f"Da xu ly truoc do: {len(processed_files)}/{len(pdf_files)} files.")

# 1. Chạy Marker cho từng file PDF (Bỏ [:3] nếu muốn chạy tất cả)
for pdf_path in pdf_files:
    file_name = os.path.basename(pdf_path)

    if file_name in processed_files:
        print(f"Bo qua (Da xu ly): {file_name}")
        continue

    print(f"Dang trich xuat: {file_name} ...")

    # Chạy Marker CLI
    result = subprocess.run(["marker_single", pdf_path, "--output_dir", OUTPUT_MD_DIR], capture_output=True)

    if result.returncode == 0:
        # Nếu thành công, cập nhật checkpoint ngay lập tức lên Drive
        processed_files.append(file_name)
        with open(CHECKPOINT_FILE, 'w') as f:
            json.dump(processed_files, f)
    else:
        print(f"Loi xu ly file {file_name}: {result.stderr.decode('utf-8')[:200]}")

print("Trích xuất Markdown hoàn tất. Đang đọc lại dữ liệu...")

# Gom tất cả các file markdown đã được sinh ra để nạp vào hệ thống
md_files = glob.glob(f"{OUTPUT_MD_DIR}/*/*.md")
for md_file in md_files:
    # Lấy tên file gốc từ tên thư mục do Marker sinh ra
    folder_name = os.path.basename(os.path.dirname(md_file))
    original_pdf_name = f"{folder_name}.pdf"

    try:
        with open(md_file, "r", encoding="utf-8") as f:
            content = f.read()
            documents.append(Document(page_content=content, metadata={"source": original_pdf_name}))
    except Exception as e:
        print(f"Loi doc file {md_file}: {e}")

# 2. Băm nhỏ văn bản (Chunking)
print("Dang bam nho van ban (Chunking)...")
text_splitter = RecursiveCharacterTextSplitter(chunk_size=800, chunk_overlap=100)
chunks = text_splitter.split_documents(documents)

# 3. Lưu vào ChromaDB
print("Dang luu vao Vector Database...")
# Nếu Database đã tồn tại, nó sẽ update thêm chunks mới
vectorstore = Chroma.from_documents(documents=chunks, embedding=embedding_model, persist_directory=VECTOR_DB_DIR)
retriever = vectorstore.as_retriever(search_kwargs={"k": 3})

print(f"Da luu {len(chunks)} chunks vao VectorDB!")
````

Cell 5: Giai đoạn 2 - Module Router (Phân luồng)

````python
from langchain_core.prompts import PromptTemplate

def intent_router(question: str) -> str:
    """Phân loại câu hỏi xem thuộc luồng Document hay API"""

    prompt = PromptTemplate(
        input_variables=["question"],
        template="""Bạn là một hệ thống phân luồng (Router).
Nhiệm vụ: Phân loại câu hỏi của người dùng vào 1 trong 2 loại sau:
1. "call_document": Nếu câu hỏi liên quan đến kiến thức, quy định, hoặc là câu hỏi trắc nghiệm A, B, C, D.
2. "call_api": Nếu câu hỏi yêu cầu thiết lập cấu hình, tạo kết nối, gọi dịch vụ hoặc điền tham số hệ thống.

Câu hỏi: {question}

Chỉ trả về chính xác 1 từ khoá: "call_document" hoặc "call_api", không giải thích gì thêm.
"""
    )

    chain = prompt | llm
    result = chain.invoke({"question": question})

    # Tiền xử lý kết quả để đảm bảo chuẩn format
    function_code = result.content.strip().replace('"', '').replace("'", "")
    if function_code not in ["call_document", "call_api"]:
        function_code = "call_document" # Default fallback

    return function_code

print("Đã khởi tạo Module Intent Router!")
````

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

# 3.1 Document RAG Agent
def document_agent(question: str, note: str) -> str:
    # 1. Nối question và note
    query = f"{question} {note}".strip()

    # 2. Tìm kiếm Context từ ChromaDB
    docs = retriever.invoke(query)
    context = "\n\n".join([d.page_content for d in docs])

    # 3. Prompt trả lời trắc nghiệm
    prompt = PromptTemplate(
        input_variables=["context", "question"],
        template="""Dựa vào tài liệu dưới đây, hãy trả lời câu hỏi trắc nghiệm.
Chỉ trả ra ký tự đáp án đúng (A, B, C, hoặc D). Không giải thích.

Tài liệu:
{context}

Câu hỏi: {question}
Đáp án:"""
    )
    chain = prompt | llm | StrOutputParser()
    answer = chain.invoke({"context": context, "question": question})
    return answer.strip()

# 3.2 API Generator Agent
def api_agent(question: str) -> str:
    # Yêu cầu LLM format chuẩn JSON
    prompt = PromptTemplate(
        input_variables=["api_data", "question"],
        template="""Dựa vào danh sách cấu hình API sau:
{api_data}

Hãy phân tích yêu cầu của người dùng và trích xuất các tham số (param) tương ứng.
Bạn PHẢI trả về kết quả dưới định dạng JSON hợp lệ, sử dụng đúng tên param chuẩn. Không có text nào khác ngoài JSON.

Yêu cầu người dùng: {question}
JSON Output:"""
    )
    chain = prompt | llm | StrOutputParser()
    answer = chain.invoke({"api_data": api_config_data, "question": question})

    # Làm sạch chuỗi trả về (Đề phòng LLM bọc trong thẻ định dạng Markdown)
    if "```json" in answer:
        answer = answer.split("```json")[1].split("```")[0].strip()
    elif "```" in answer:
        answer = answer.split("```")[1].split("```")[0].strip()

    return answer.strip()

print("Đã khởi tạo RAG Agent và API Agent!")
````

Cell 6: Giai đoạn 3 - Các Agent Xử lý Lõi

````python
import time

def process_chatbot_request(item_id: str, question: str, note: str = "") -> dict:
    """Luồng xử lý tổng: Nhận Input -> Định tuyến -> Xử lý -> Trả Output"""
    start_time = time.time()

    # 1. Routing
    function_code = intent_router(question)

    # 2. Rẽ nhánh xử lý
    function_answer = ""
    if function_code == "call_document":
        function_answer = document_agent(question, note)
    elif function_code == "call_api":
        function_answer = api_agent(question)

    # 3. Tính thời gian xử lý
    time_response = round(time.time() - start_time, 2)

    # 4. Đóng gói JSON
    output = {
        "id": item_id,
        "function_code": function_code,
        "function_answer": function_answer,
        "time_response": f"{time_response}s"
    }

    return output
````

Cell 7: Giai đoạn 4 - Orchestration (Tích hợp luồng Main)

````python
def process_chatbot_request(item_id: str, question: str, note: str = "") -> dict:
    """Luồng xử lý tổng: Nhận Input -> Định tuyến -> Xử lý -> Trả Output"""
    start_time = time.time()

    # 1. Routing
    function_code = intent_router(question)

    # 2. Rẽ nhánh xử lý
    function_answer = ""
    if function_code == "call_document":
        function_answer = document_agent(question, note)
    elif function_code == "call_api":
        function_answer = api_agent(question) # Không truyền note

    # 3. Tính thời gian xử lý
    time_response = round(time.time() - start_time, 2)

    # 4. Đóng gói JSON
    output = {
        "id": item_id,
        "function_code": function_code,
        "function_answer": function_answer,
        "time_response": f"{time_response}s"
    }

    return output

print("Đã khởi tạo luồng Main Orchestration!")
````
